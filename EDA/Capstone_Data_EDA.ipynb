```python
grouped_data = df.groupby(df['If yes, please specify your salary range per annum(in Rupee)?'])
```

```python
grouped_data_corr = grouped_data.apply(lambda group: group.corr())
grouped_data_corr
```

This correlations can be used as weights in the optimiser

```python
sns.heatmap(grouped_data.get_group('Below 3,00,000').corr(),annot = True)
```
